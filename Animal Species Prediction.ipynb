```python
# Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions, VGG16
from IPython.display import Image, display

# Load the pre-trained VGG16 model with ImageNet weights
vgg_model = VGG16(weights='imagenet')
```

```python
def preprocess_image(image_path):
    # Load the image and preprocess it for the VGG16 model
    img = load_img(image_path, target_size=(224, 224))
    img_array = img_to_array(img)
    img_array = preprocess_input(img_array)
    img_array = np.expand_dims(img_array, axis=0)
    
    return img_array
```

```python
def predict_top_species(image_path, top_n=5):
    # Preprocess the image and predict top N species
    img_array = preprocess_image(image_path)
    preds = vgg_model.predict(img_array)
    decoded_preds = decode_predictions(preds, top=top_n)[0]
    
    species_list = []
    for (_, species, prob) in decoded_preds:
        species_list.append((species, prob))
    
    return species_list
```

```python
def display_species_and_image(species_list, image_path):
    # Display top predicted species
    print('Top predicted animal species:')
    for idx, (species, prob) in enumerate(species_list, start=1):
        print(f'{idx}. {species}: {prob:.2f}')
    
    # Display the input image
    display(Image(filename=image_path))
```

```python
def main():
    # Specify the path to the image
    image_path = 'bird.jpg'
    
    # Predict top species and display results
    top_species_list = predict_top_species(image_path)
    display_species_and_image(top_species_list, image_path)

if __name__ == "__main__":
    main()
```
